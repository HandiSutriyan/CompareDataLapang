```python
import numpy as np
import pandas as pd
```

```python
df_uut = pd.read_csv('./datasets/awosdata-21-22-CS Aero.csv',sep=None, engine='python')
df_uut
```

```python
df_uut['Time (UTC)'] = pd.to_datetime(df_uut['Time (UTC)'])
```

```python
df_uut.rename(columns={'Time (UTC)':'Timestamp'},inplace=True)
df_uut.head()
```

```python
df_uut_resampled = df_uut.set_index("Timestamp").resample("1min").mean().reset_index()
df_uut_resampled.head()
```

```python
df_uut_resampled.to_csv('data-awos-resamples.csv',index=False)
```
